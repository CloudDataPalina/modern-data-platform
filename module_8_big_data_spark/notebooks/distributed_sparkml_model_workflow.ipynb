# 5. Create a Spark DataFrame with Sample Data

A structured dataset is created to simulate a Machine Learning regression workflow using Apache Spark.

The dataset contains:
- infant height measurements

- numerical features for regression analysis

The dataset is converted into a Spark DataFrame for distributed processing and SparkML model training.

```python
# Create sample dataset
mydata = [[46,2.5],[51,3.4],[54,4.4],[57,5.1],[60,5.6],[61,6.1],[63,6.4]]

# Define columns
columns = ["height", "weight"]

# Create Spark DataFrame
mydf = spark.createDataFrame(mydata, columns)

# Display dataset
mydf.show()
```

# 6. Feature Engineering with VectorAssembler

Feature engineering is performed using the `VectorAssembler` class to transform numerical input columns into Spark ML feature vectors.

This transformation is required because SparkML algorithms operate on vectorized feature representations.

The workflow includes:
- feature column transformation
- vector creation
- ML-compatible dataset preparation
- structured feature engineering for regression modeling

```python
assembler = VectorAssembler(
    inputCols=["height"],
    outputCol="features"
)

data = assembler.transform(mydf).select('features', 'weight')

data.show()
```

## Feature Engineering Results

The VectorAssembler successfully transformed the input feature columns into Spark ML feature vectors.

Key observations:
- numerical height values were successfully converted into vectorized features
- 7 records were transformed into SparkML-compatible feature vectors
- the dataset structure is compatible with distributed SparkML algorithms
- the transformed DataFrame is ready for regression model training
- feature vectors were successfully generated for distributed ML processing

Generated dataset structure:
- feature column: `features`
- target column: `weight`
- total transformed records: 7
- sample feature vector: `[46.0]`
- sample target value: `2.5`

The feature engineering stage successfully prepared the dataset for distributed Machine Learning processing.

# 7. Train the Linear Regression Model

A Linear Regression model is created and trained using Spark MLlib to simulate a distributed Machine Learning workflow.

The training process includes:
- regression model initialization
- parameter configuration
- distributed model training
- predictive relationship learning between height and weight values

The model is trained using SparkML distributed processing capabilities.

```python
# Create Linear Regression model
lr = LinearRegression(featuresCol='features', labelCol='weight', maxIter=100)

lr.setRegParam(0.1)

# Train model
lrModel = lr.fit(data)
```

# 10. Execute Distributed Predictions

The loaded SparkML model is used to generate predictions for new input values using distributed Spark inference workflows.

The prediction pipeline includes:
- feature vector generation
- DataFrame transformation
- distributed inference execution
- regression prediction generation

The workflow simulates enterprise predictive analytics execution using Apache Spark.

```python
# Prediction function
def predict(height):
    assembler = VectorAssembler(
        inputCols=["height"],
        outputCol="features"
    )

    data = [[height, 0]]
    columns = ["height", "weight"]

    df = spark.createDataFrame(data, columns)

    transformed_df = assembler.transform(df).select('features', 'weight')

    predictions = model.transform(transformed_df)

    predictions.select("prediction").show()
```

# 11. Predict Infant Weight for Height = 70 cm

A distributed prediction is executed using the pretrained SparkML regression model.

The model generates an estimated weight prediction based on the provided height input value.

```python
predict(70)
```
