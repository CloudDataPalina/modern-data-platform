## 4. Search Term Analytics

Distributed analytical queries are executed on the Spark DataFrame to identify search behavior patterns and analyze the frequency of search terms within the e-commerce dataset.

### Search Frequency Analysis

The dataset is filtered to determine how many times a specific search term was queried by users.

```python
from pyspark.sql.functions import count

df.filter(df.searchterm == "gaming laptop").count()
```

### Search Query Results

The term `gaming laptop` appeared 499 times in the dataset.

This indicates that gaming-related product searches represent a significant portion of user search activity within the analyzed e-commerce platform.

### Top Search Terms Analysis

Aggregation operations are executed using Spark DataFrames to identify the most frequently searched product terms.

```python
from pyspark.sql.functions import count

df.groupby("searchterm") \
    .agg(count("*").alias("count")) \
    .orderBy("count", ascending=False) \
    .show(5)
```

### Analytical Results

The analysis identified the most frequently searched product categories and device types.

Key observations:
- mobile device searches dominate the dataset
- `mobile 6 inch` and `mobile 5g` are the most frequently searched terms
- smartphone-related searches appear significantly more often than laptop or tablet searches
- search behavior indicates strong user interest in mobile technology products

The Spark aggregation workflow successfully demonstrated distributed analytical processing using Spark DataFrames.

```python
from pyspark.ml.regression import LinearRegressionModel

# Load pretrained SparkML model from local repository
model = LinearRegressionModel.load(
    "../models/sales_prediction.model"
)
```

## 6. Distributed ML Inference and Sales Forecasting

The pretrained Spark ML regression model is used to generate distributed sales forecasting predictions for future periods.

A Spark DataFrame containing future year values is created and transformed into feature vectors for distributed ML inference execution.

### Prepare Forecasting Dataset

A prediction dataset is created for future sales forecasting using the year 2023 as input for the regression model.

```python
from pyspark.ml.feature import VectorAssembler

# Create forecasting dataset
data = [[2023]]
columns = ["year"]

df_2023 = spark.createDataFrame(data, columns)

# Transform dataset into feature vectors
assembler = VectorAssembler(
    inputCols=["year"],
    outputCol="features"
)

final_data = assembler.transform(df_2023)
```

### Feature Engineering Results

The forecasting dataset was successfully transformed into Spark ML feature vectors.

Key observations:
- the future prediction dataset contains year-based forecasting input
- Spark ML feature engineering was successfully executed
- the dataset is now prepared for distributed prediction processing

### Execute Distributed Predictions

The pretrained Spark ML model is executed on the forecasting dataset to generate future sales predictions.

```python
# Execute distributed predictions
predictions = model.transform(final_data)

# Display prediction results
predictions.select("prediction").show()
```

### Forecasting Results

The distributed Spark ML inference workflow successfully generated future sales forecasting predictions.

Prediction output:
- forecasted sales value for 2023: approximately 175.17

Key observations:
- the pretrained regression model successfully executed distributed inference
- Apache Spark handled predictive analytics processing within the distributed environment
- the forecasting workflow demonstrated scalable ML execution using Spark ML
